```python
#Метод среднего:
#Статистика до обработки выбросов:
mean_before = df['salary'].mean()
std_before = df['salary'].std()
print(df.columns)
#Обработка выбросов:
# Определение границ для выбросов (например, с использованием z-оценки)
z_score_threshold = 3
df_no_outliers_mean = df[(df['salary'] - df['salary'].mean()).abs() < z_score_threshold * 3 * df['salary'].std()]
print(df.columns)
#Статистика после обработки выбросов:
mean_after = df_no_outliers_mean['salary'].mean()
std_after = df_no_outliers_mean['salary'].std()
print(df.columns)
```

```python
#Межквартильный размах (IQR):
#Статистика до обработки выбросов:
q1 = df['salary'].quantile(0.25)
q3 = df['salary'].quantile(0.75)
iqr_before = q3 - q1
print(df.columns)
#Обработка выбросов:
lower_bound = q1 - 1.5 * iqr_before
upper_bound = q3 + 1.5 * iqr_before
df_no_outliers_iqr = df[(df['salary'] >= lower_bound) & (df['salary'] <= upper_bound)]
print(df.columns)
#Статистика после обработки выбросов:
q1_after = df_no_outliers_iqr['salary'].quantile(0.25)
q3_after = df_no_outliers_iqr['salary'].quantile(0.75)
iqr_after = q3_after - q1_after
print(df.columns)
```

Вывод статистики и объяснение:
Сравнение статистик до и после обработки выбросов позволяет оценить влияние выбросов на распределение данных.

Метод среднего:

Если среднее значение существенно изменяется после удаления выбросов, это может указывать на наличие значительных выбросов в данных.
Стандартное отклонение также может измениться, что отразит уровень изменчивости данных.
Межквартильный размах (IQR):

Используется для выявления выбросов, основываясь на разнице между первым и третьим квартилями.
Если границы выбросов (lower_bound и upper_bound) сильно отличаются от квартилей, это может указывать на наличие выбросов.
Объяснение результатов может включать в себя анализ причин появления выбросов, оценку их влияния на статистики, и принятие решения о том, следует ли их удалить или оставить в данных в зависимости от контекста задачи и характера данных.

##Дубликаты

Проверьте датафрейм на дубликаты и удалите их, если дублей нет - создайте.

```python
import pandas as pd

# Предположим, что «df» — это ваш DataFrame.
# Проверка дубликатов
duplicates_exist = df.duplicated().any()#Эта строка проверяет наличие дубликатов в DataFrame. Метод .duulated() возвращает логическую серию, указывающую, является ли каждая строка дубликатом или нет. Метод .any() проверяет, существует ли хотя бы одно значение True, что означает наличие дубликатов.

# Удаление дубликатов, если они существуют.
if duplicates_exist:
    df_no_duplicates = df.drop_duplicates()
    print("Дубликаты удалены. Новый DataFrame без дубликатов:")
    print(df_no_duplicates)#Этот код удаляет дубликаты с помощью метода .drop_duplications() и печатает новый DataFrame без дубликатов.
else:
    #Создание дубликатов, если их нет
    df_with_duplicates = df.append(df)  # Добавление DataFrame к самому себе для создания дубликатов
    print("No duplicates found. New DataFrame with duplicates created:")
    print(df_with_duplicates)#Этот код добавляет DataFrame к самому себе, создавая дубликаты, а затем печатает новый DataFrame с дубликатами.
```

## Преобразование признаков

Масштабируйте числовые признаки через MinMaxScaler, StandardScaler

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder

# Предположим, что «df» — это ваш DataFrame как с числовыми, так и с категориальными функциями.

# Отдельные числовые и категориальные столбцы
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Создание экземпляров масштабаторов
min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
encoder = OneHotEncoder()

# Масштабирование числовых характеристик
df_scaled_numerical = df.copy()
df_scaled_numerical[numerical_columns] = standard_scaler.fit_transform(df[numerical_columns])

# Категориальные функции горячего кодирования
df_encoded_categorical = pd.get_dummies(df[categorical_columns])

# Объединение масштабированных числовых и категориальных функций с горячим кодированием
df_final = pd.concat([df_scaled_numerical, df_encoded_categorical], axis=1)

# Распечатайте окончательный DataFrame
print("Final DataFrame:")
print(df_final)
```

Этот код представляет собой пример использования библиотеки pandas и scikit-learn для предобработки данных в DataFrame, включающем как числовые, так и категориальные признаки. Давайте разберем каждый шаг:

1. `numerical_columns` и `categorical_columns`: Эти две строки кода идентифицируют числовые и категориальные столбцы в вашем DataFrame `df` с использованием метода `select_dtypes`.

2. Создание экземпляров масштабаторов и кодировщика:
   - `min_max_scaler`: Экземпляр MinMaxScaler, который используется для масштабирования числовых признаков от 0 до 1.
   - `standard_scaler`: Экземпляр StandardScaler, который используется для стандартизации числовых признаков (среднее = 0, стандартное отклонение = 1).
   - `encoder`: Экземпляр OneHotEncoder, который используется для горячего кодирования категориальных признаков.

3. `df_scaled_numerical`: Создание копии DataFrame `df` для числовых признаков и применение стандартизации с использованием `standard_scaler` к числовым столбцам.

4. `df_encoded_categorical`: Горячее кодирование категориальных признаков с использованием `pd.get_dummies`. Эта функция создает новые бинарные столбцы для каждой уникальной категории в категориальных столбцах.

5. `df_final`: Объединение DataFrame `df_scaled_numerical` (с числовыми признаками, подвергшимися стандартизации) и `df_encoded_categorical` (с горячим кодированием категориальных признаков) по оси 1.

6. Наконец, код печатает окончательный DataFrame `df_final`.

```python
import pandas as pd
import numpy as np


def corr_pearson(column_1, column_2):
  cov = np.cov(column_1, column_2)
  return cov/(column_1.std() * column_2.std())
```

```python
data = {
    'Рост': [170, 155, 180, 162, 175],
    'Вес': [65, 50, 75, 55, 70]
}
df = pd.DataFrame(data)
df
```
