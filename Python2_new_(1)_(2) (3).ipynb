*p.s. Если у вас только один числовой признак, преобразуйте категориальный при помощи LabelEncoder (https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html)*

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Замените 'ds_salaries.csv' на реальное имя вашего файла
df = pd.read_csv('ds_salaries.csv')

# Предположим, у вас есть числовые и категориальные признаки, например, 'work_year' и 'salary'
# Если у вас только один числовой признак, преобразуйте категориальный при помощи LabelEncoder
# Например, преобразуем 'experience_level' с помощью LabelEncoder
label_encoder = LabelEncoder()
df['experience_level_encoded'] = label_encoder.fit_transform(df['experience_level'])

# Построим диаграмму рассеивания для числовых признаков
plt.figure(figsize=(10, 8))
sns.scatterplot(x='work_year', y='salary', data=df, hue='experience_level_encoded', palette='viridis')
plt.title('Диаграмма рассеивания для числовых признаков')
plt.xlabel('Год работы')
plt.ylabel('Заработная плата')
plt.show()
```

```python
# Вычисление корреляционной матрицы
correlation_matrix = df.corr()

# Визуализация корреляции
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
